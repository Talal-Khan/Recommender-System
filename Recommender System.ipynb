```python
moviemat=df.pivot_table(index='userId',columns='title',values='rating')
```

```python
moviemat
```

```python
ratings.sort_values('Num of Rating',ascending=False).head(10)
```

```python
forrestgump_user_rating=moviemat['Forrest Gump (1994)']
shawshank_user_rating=moviemat['Shawshank Redemption, The (1994)']
```

```python
forrestgump_user_rating.head()
```

```python
shawshank_user_rating.head()
```

```python
similar_to_forrestgump=moviemat.corrwith(forrestgump_user_rating)
similar_to_shawshank=moviemat.corrwith(shawshank_user_rating)
```

```python
corr_forrestgump=pd.DataFrame(similar_to_forrestgump,columns=['Correlation'])
corr_forrestgump.dropna(inplace=True)
```

```python
corr_forrestgump.head()
```

```python
corr_forrestgump.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_forrestgump=corr_forrestgump.join(ratings['Num of Rating'])
```

```python
corr_forrestgump.head()
```

```python
corr_forrestgump[corr_forrestgump['Num of Rating']>100].sort_values('Correlation',ascending=False)
```

```python
corr_shawshank=pd.DataFrame(similar_to_shawshank,columns=['Correlation'])
corr_shawshank.dropna(inplace=True)
```

```python
corr_shawshank.head()
```

```python
corr_shawshank.sort_values('Correlation',ascending=False)
```

```python
corr_shawshank=corr_shawshank.join(ratings['Num of Rating'])
```

```python
corr_shawshank.head()
```

```python
corr_shawshank[corr_shawshank['Num of Rating']>100].sort_values('Correlation',ascending=False).head(10)
```
